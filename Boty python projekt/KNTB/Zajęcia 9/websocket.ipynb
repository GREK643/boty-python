```python
import websocket
from binance.exceptions import BinanceAPIException
from binance.client import Client
from binance.enums import * 
import json
import time
import hmac
import hashlib
import requests
from urllib.parse import urljoin, urlencode
import pandas as pd
import pprint
```

```python

socket = 'wss://stream.binance.com:9443/ws'



def on_open(ws):
  subscribe_message = {"method": "SUBSCRIBE", "params":["dogeusdt@ticker"],"id": 1}
  ws.send(json.dumps(subscribe_message))


def on_message(ws, message):
  print("received a message")
  print(json.loads(message))

def on_close(ws):
  print("closed connection")

ws = websocket.WebSocketApp(socket, on_open=on_open, on_message=on_message, on_close=on_close)
ws.run_forever()
```

```python
client = Client('','')
tic = client.get_all_tickers()

list_of_coins = []

for x in tic:
  list_of_coins.append(x['symbol'])
```

```python
import requests
import time

start_time = time.time()

for coin in list_of_coins[:100]:
    url = f'https://api.binance.com/api/v3/ticker/price?symbol={coin}'
    resp = requests.get(url)
    symbol = resp.json()
    print(symbol['symbol'],symbol['price'] )

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import aiohttp
import asyncio
import time

start_time = time.time()


async def main():

    async with aiohttp.ClientSession() as session:

        for coin in list_of_coins[:100]:
            url = f'https://api.binance.com/api/v3/ticker/price?symbol={coin}'
            async with session.get(url) as resp:
                symbol = await resp.json()
                print(symbol['symbol'])

asyncio.run(main())
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import aiohttp
import asyncio
import time

start_time = time.time()


async def get_ticker(session, url):
    async with session.get(url) as resp:
        coins = await resp.json()
        return coins['symbol']


async def main():

    async with aiohttp.ClientSession() as session:

        tasks = []
        for coin in list_of_coins[:100]:
            url = f'https://api.binance.com/api/v3/ticker/price?symbol={coin}'
            tasks.append(asyncio.ensure_future(get_ticker(session, url)))

        original_ticker = await asyncio.gather(*tasks)
        for t in original_ticker:
            print(t)

asyncio.run(main())
print("--- %s seconds ---" % (time.time() - start_time))

```
