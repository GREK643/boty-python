```
def intervals():
  df['1H'] = df['close'].pct_change(periods=60)*100
  df['24H'] = df['close'].pct_change(periods=1440)*100
  df['1W'] = df['close'].pct_change(periods=10080)*100  
```

```
def symulacja_SL2(daily_change=20, weekly_change=20,difference=10, pp=36, name='BTCUSDT', cash=25):
  
  i=60*24*7
  zysk=0
  kupione = 0
  max_price=np.nan
  profit = 0
  
  
  while i< len(df):
    if max_price < df['close'][i]:
      max_price = df['close'][i]

    if df['24H'][i]>=daily_change and df['1W'][i]>=weekly_change and kupione==0 and abs(df['24H'][i]-df['1W'][i])<=difference:
      kupno=df['close'][i]
      ile_kupic = cash/kupno
      kupione=1
      max_price = df['close'][i]
      
    
    if kupione==1 and df['close'][i] < (100-pp)*0.01*max_price:
      sell = (df['close'][i])*ile_kupic
      zysk=sell-cash
      ile_kupic=0
      kupno=0
      kupione = 0
      profit += zysk
      
    i+=1
  return profit
   
def append(name, start=20, end=40):

  wyniki=[]
  for i in range(start,end):
    profit = symulacja_SL2(pp=i, daily_change=15, weekly_change=15,difference=5)
    wyniki.append([profit,i,name])
    print(i, profit)
  return wyniki
```

```
df = pd.read_csv('/content/drive/MyDrive/coiny/1INCHUSDT.csv')
intervals()
lista = append('1INCHUSDT')
tab = pd.DataFrame(lista, columns=['Profit', 'SL', 'Coin']) 
tab
```

```
main_df = pd.DataFrame()
for coin in ['LSKUSDT', 'BTCUSDT']:

  df = pd.read_csv('/content/drive/MyDrive/coiny/{}'.format(coin)+'.csv')
  intervals()
  lista = append(name=coin, start=20, end=25)
  tab = pd.DataFrame(lista, columns=['Profit', 'SL', 'Coin']) 
  main_df = pd.concat([main_df, tab])
```
